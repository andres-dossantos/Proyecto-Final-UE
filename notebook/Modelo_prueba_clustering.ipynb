```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
## Este ejercicio toma la data de "data_total_countries_xcolumna" para un modelo de clustering k-means
## Cargar el archivo data_total_countries 1.csv
country_df = pd.read_csv('data_total_country_xcolumna.csv')
country_df.head()
```

```python
##country_df
# Crear un nuevo DataFrame para almacenar los valores más recientes de cada indicador
new_df_ind = country_df[['Country Name', 'Region', 'Continent']].copy()

# Obtener la lista de indicadores únicos
indicators = set(col.split('_')[0] for col in country_df.columns if '_' in col)

# Iterar sobre cada indicador
for indicator in indicators:
    # Filtrar las columnas correspondientes al indicador actual
    indicator_cols = [col for col in country_df.columns if col.startswith(indicator)]
    
    # Encontrar el valor más reciente para cada país
    new_df_ind[indicator] = country_df[indicator_cols].bfill(axis=1).iloc[:, 0]

# Mostrar el DataFrame filtrado con los valores más recientes de cada indicador
new_df_ind.head()
```

```python
# Guardar el DataFrame filtrado en un archivo CSV
#new_df_ind.to_csv('filtered_data.csv', index=False)
```

```python
# Filtrar la columna "Continent" para incluir solo "Europe" y "America"
filtered_df = new_df_ind[new_df_ind['Continent'].isin(['Europe', 'South America', 'North America'])]
```

```python
# Contar el número de valores faltantes en cada columna
conteo_nulos = filtered_df.isnull().sum()
print(conteo_nulos)
```

```python
#Columnas a eliminar
cols_to_drop = ['Gasto investigación y desarrollo',
                'Tecnicos investigación y desarrollo',
                'Tasa crecimiento prom. anualizada ',
                'Cobertura de beneficios por desempleo y los ALMP ',
                'Exportaciones bienes y servicios (% del PIB)',
                'Exportaciones de bienes y servicios(% anual)',
                'Trabajadores salud']
filtered_df = filtered_df.drop(cols_to_drop, axis=1) ##  
```

```python
# Reemplazar los valores nulos por cero en todo el dataset
filtered_df.fillna(0, inplace=True)
```

```python
# Seleccionar las variables relevantes
features = [

    
    
    'Crecimiento población urbana',
    'Inflación, deflactor del PIB (%) anual',
    'Personas desempleadas educación intermedia',
    'Desempleo hombres',
    'Desempleo mujeres', 
    'Empleados agricultura - hombres',
    'Comercio servicios', 
    'Crecimiento PIB per capita',
    'Contaminación aire', 
    'Trabajadores asalariados (empleados), total',
    'Indice facilidad para hacer negocios',
    'Personas desempleadas educación basica', 
    'Empleadores, total',
    'Agricultura, silvicultura y pesca, valor agregado',
    'Tierras cultivables', 
    'Personas desempleadas educación avanzada',
    'PIB per capita US a precios actuales', 
    'Inversión extranjera directa',
    'Empleados agricultura - mujeres', 
    'Empleo vulnerable, total',
    'Empleos agricultura',
    'Desempleo, total', 
    'Pago de impuestos (numero)']
    
    
    
    
    
    
    
    
    
    
]
X = filtered_df[features]
```

```python
# Normalizar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Determinar el número óptimo de clusters usando el método del codo
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Método del Codo')
plt.xlabel('Número de Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Calcular el Silhouette Score para diferentes números de clusters
silhouette_scores = []
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    cluster_labels = kmeans.fit_predict(X_scaled)
    silhouette_avg = silhouette_score(X_scaled, cluster_labels)
    silhouette_scores.append(silhouette_avg)

plt.plot(range(2, 11), silhouette_scores)
plt.title('Silhouette Score')
plt.xlabel('Número de Clusters')
plt.ylabel('Silhouette Score')
plt.show()

# Seleccionar el número óptimo de clusters basado en el Silhouette Score más alto
optimal_clusters = silhouette_scores.index(max(silhouette_scores)) + 2
print(f"El número óptimo de clusters basado en el Silhouette Score es: {optimal_clusters}")
```

```python
# Aplicar K-Means con el número óptimo de clusters (por ejemplo, k=3)
kmeans = KMeans(n_clusters=7, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Añadir los clusters al DataFrame original
filtered_df['Cluster'] = clusters
```

```python
# Guardar el DataFrame con los clusters
filtered_df.to_csv('clustered_countries_v2.csv', index=False)
```
